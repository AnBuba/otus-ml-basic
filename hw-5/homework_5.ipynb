```python
# Проверка пропусков (Nan)
data.isna().sum()
```

```python
# Удаление столбца Unnamed: 32, так как в нем все Nan
data = data.drop(columns = ['Unnamed: 32'],axis = 1)
```

```python
data
```

```python
# При построении ROC-кривой столкнулась с проблемой, что диагноз обозначается как M и B (строкойвый тип).
# Решила заменить M на 1 и B на 0

data = data.replace(['M'],1)
data = data.replace(['B'],0)
```

```python
data
```

```python
# Разбейте данные на train-test, отложив 30% выборки для тестирования.
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data.drop('diagnosis', axis=1), data.diagnosis, \
                                                    test_size=0.3, random_state=42)
```

```python
X_train
```

```python
y_train
```

```python
# Приведите все непрерывные переменные к одному масштабу при помощи стандартизации
# Масштабирование необхоимо из-за того, что признаки измеряются в разных единицах. 
# И масштабирование помогает не потерять информацию о каждой из точек данных. 
scaler = StandardScaler()

scaler.fit(X_train)
X_train = scaler.transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
X_train
```

```python
# Постройте модель kNN
from sklearn.neighbors import KNeighborsClassifier

classifier = KNeighborsClassifier(n_neighbors=5)
classifier.fit(X_train, y_train)

y_pred = classifier.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test, y_pred))
```

```python
print(classification_report(y_test, y_pred))
```

```python
# Точность модели на полном тестовом наборе
classifier.score(X_test, y_test)
```
